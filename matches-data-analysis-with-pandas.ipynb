# Handling Missing Values :

## 1)Droping missing data

```python
matches.isnull().sum()
```

### there are 7 null values in city column 3 in winner and 1 in umpire 1 and 2 636 in umpire 3 i.e all the values in umpire 3 is missing 
dropna could be done on column where column will be deleted or on rows where row will be deleted

### axis-1 for column,0- row, default is 0
#### how - any : delete the row or column if there will be missing value in any one of the row or colun.... all:will delete if and only if all the values of particular row or column will be null
#### use inplace parameter for permanent changes
#### subset : pass columns name inside the list if there will be missing value in that column then deletion will be executed

```python
matches.dropna(how='all',axis=1,inplace=True)
matches
```

```python
matches.dropna(subset=['umpire1','umpire2']).shape
```

#### we can see 1 row is deleted as there was 1 missing value

```python
matches.fillna(0)
```

### all missing value filled by 0
